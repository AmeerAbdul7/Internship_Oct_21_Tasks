### <font color = brown>Variance: 

<ol>
<ul> 
<font color =navy><li>Variance measures how far are data points spread out from the mean.<li>A high variance indicates that data points are spread widely and a small variance indicates that the data points are closer to the mean of the data set.

### <font color = brown>Formula:

<font color = navy>$$ \sigma^2 = \left\{\frac{\Sigma^n_{i=1}|\mu-obj_i|}{N}\right\}$$
    Where <li>**N** is no of Observations, <li>**obj** is individual Datapoint from a Data Set, and <li>mean is**$$\mu$$**.

### <font color =darkorange> Example from Scratch and by using Libraries

```python
def vrnc(x):
    mean_ = sum(x) / len(x)
    dev = [(obj_i - mean_) ** 2 for obj_i in x] 
    var = sum(dev) / len(x)
    return var

x = df['Annual_HH_Income']
vrnc(x)
```

```python
Anl_incm_variance = st.pvariance(df['Annual_HH_Income'])
Anl_incm_variance 
```

<ol>
<ul> 
<font color =navy><li><b>Note</b>: that this implementation of ddof which defaults to 0.<li> This argument allows us to set the <b>degrees of freedom</b> that we want to use when calculating the variance. <li>For example, ddof=0 will allow us to calculate the variance of a <b>population</b>. <li>Meanwhile, <b>ddof=1</b> will allow us to estimate the population variance using a <b>sample of data</b>.
    <li>If import the Library as <b>statistics</b> module and then call <b>pvariance()</b> with our data as an argument. <li>That will return the variance of the <b>population</b>.

<li>On the other hand, we can use <b>Python's variance()</b> to calculate the variance of a <b>sample</b>. <li>That's because variance() uses <b>n - 1</b> instead of <b>n</b> to calculate the variance. 

```python
def vrnc(x,ddof=0):
    mean_ = sum(x) / len(x)
    dev = [(obj_i - mean_) ** 2 for obj_i in x] 
    var = sum(dev) / (len(x)-ddof)
    return var

x = df['Annual_HH_Income']
vrnc(x,ddof=1)
```

<ol>

    <li>It gives Distribution of Variables

    <li> <b>Note</b>: <b>Range</b> is <b>Effected</b> when Outliers Imputed.

### <font color = brown>Formula:

<li> <font color=navy>Simply we Formulaed Standard Deviation over population as $$ (\sigma) = \sqrt{Variance} \hspace{0.5cm}\approx \hspace{0.5cm} \sqrt{\left\{\frac{\Sigma^n_{i=1}|\mu-obj_i|}{N}\right\}}$$
    <li> <font color=navy> Standard Deviation over Sample as $$ (S) = \sqrt{Variance} \hspace{0.5cm}\approx \hspace{0.5cm} \sqrt{\left\{\frac{\Sigma^n_{i=1}|\mu-obj_i|}{N-1}\right\}}$$


### <font color =darkorange> Example from Scratch and by using Libraries

```python
# Std of Papulation

def std_(x):
    mean_ = sum(x) / len(x)
    dev = [(obj_i - mean_) ** 2 for obj_i in x] 
    var = sum(dev) / len(x)
    stdev = np.sqrt(var)
    return stdev

x = df['Annual_HH_Income']
std_(x)
```

### <font color = brown>Normal Distribution: 

<ol>
<ul> 
<font color =navy>
<li>Normal distribution is a probability distribution that is symmetric about the mean.<li>showing that data near the mean are more frequent in occurrence than data far from the mean.<li> In graph form, normal distribution will appear as a <b>bell curve.</b><li> it also known as the <b>Gaussian distribution</b>.
    <li>statistically normal distribution has the mean as its <b>mean</b>, its <b>mode</b>, and its <b>median</b> and it looks like a <b>symmetric mound or bell.</b>
        <li>the word normal is a very powerful adjective when used to describe a frequency distribution or when used to describe the data of a sample or population. 



<font color = navy>The Data Is Centered About the <b>Mean-Mode-Median</b>
<ol>
<ul> 
<font color = navy><b>Empirical rule</b>
<li>The empirical rule, or the 68-95-99.7 rule, tells us where most of values lie in a normal distribution:
<li>Around 68% of values are within 1 standard deviation from the mean.
<li>Around 95% of values are within 2 standard deviations from the mean.
<li>Around 99.7% of values are within 3 standard deviations from the mean.

    
    
    




### <font color = brown>Feature of Normal Distribution: 

<ol>
<ul>    
   <font color = navy><li> The mean-mode-median is in the center.
<li>It is the mean because it is the ARITHMETIC average of all the scores.
<li>It is the mode because of all the scores the mean score happens MOST often.
<li>It is the median because when the scores are displayed from lowest to highest, the mean is the MIDDLE score, the median.
<li>The EXPECTED value is the mean.
   <li> The frequency curve is bell shaped.
<li>The bell shape has perfect bilateral symmetry - the left balances exactly with the right.
<li>The score at -2 is balanced by a score at +2 and the frequencies from 0 to +2 and from 0 to -2 are equal.
<li>The area under the curve from 0 to +2 is exactly the same as the area under the curve from 0 to -2.
<li>Fifty percent of the scores are above the mean and 50% are below the mean.
<li>The probability a score is above the mean is 50% and the probability a score is below the mean is 50%.
    <li><font color = navy>Most of the scores are in<b> the middle, about the mean, and few are in the tails, at the extremes.</b>



        
#### <font color = black>Axiom's of Normal Distribution:    
<li>The area under the curve is equal to 1.
<li>The probability of an event that does not happen is 0.
<li>The sum of the probabilities of all events is 1.
<li>The standard deviation tells one how the scores are spread out and therefore the fatness or skinniness of the bell.

### <font color = brown>Formula:

<ol>
<ul>    
   <font color = navy><li>There are several ways to measure skewness. Pearson’s first and second coefficients of skewness are two common ones.<li> Pearson’s first coefficient of skewness, or Pearson mode skewness, subtracts the mode from the mean and divides the difference by the standard deviation.<li> Pearson’s second coefficient of skewness, or Pearson median skewness, subtracts the median from the mean, multiplies the difference by three, and divides the product by the standard deviation.
       
<li>The Formulae for Pearson's Skewness Are:
       
 $$ Sk_1\hspace{0.3cm} = \hspace{0.3cm}\frac {\mu - Mo}{s}$$
 $$ Sk_2 \hspace{0.3cm}= \hspace{0.3cm}\frac {3\mu - Md}{s}$$
    

<li><b>Where........</b><li>Sk_1 is the Pearson’s first coefficient of skewness and Sk2 is the second
<li>s=the standard deviation for the sample

<li>\mu =is the mean value
<li>Mo=the modal (mode) value
<li>Md=is the median value

### <font color = brown>Explaination of Quantile-Quantile(QQ) Plot: 

<ol>
<ul>    
   <font color = navy><li>The Q-Q plot, or quantile-quantile plot, is a graphical tool to help us assess if a set of data plausibly came from some theoretical distribution such as a Normal or exponential.<li> For example, if we run a statistical analysis that assumes our dependent variable is Normally distributed, we can use a Normal Q-Q plot to check that assumption. <li>It’s just a visual check, not an air-tight proof, so it is somewhat subjective. <li>But it allows us to see at-a-glance if our assumption is plausible, and if not, how the assumption is violated and what data points contribute to the violation.
       <li>A Q-Q plot is a scatterplot created by plotting two sets of quantiles against one another.<li>If both sets of quantiles came from the same distribution, we should see the points forming a line that’s roughly straight.
       <li>These are often referred to as “percentiles”. <li>These are points in your data below which a certain proportion of your data fall.<li> For example, imagine the classic bell-curve standard Normal distribution with a mean of 0.<li> The 0.5 quantile, or 50th percentile, is 0.<li> Half the data lie below 0. That’s the peak of the hump in the curve. <li>The 0.95 quantile, or 95th percentile, is about 1.64.<li> 95 percent of the data lie below 1.64.

### <font color = brown>Implementation of Box Cox::

```python
data = df.Annual_HH_Income
  
# transform training data & save lambda value
fitted_data, fitted_lambda = stats.boxcox(data)
  
# creating axes to draw plots
fig, ax = plt.subplots(1, 2)
  
# plotting the original data(non-normal) and 
# fitted data (normal)
sns.distplot(data, label = "Non-Normal", ax = ax[0])
  
sns.distplot(fitted_data, label = "Normal", color ="green", ax = ax[1])
  
# adding legends to the subplots
plt.legend(loc = "upper right")
  
# rescaling the subplots
fig.set_figheight(5)
fig.set_figwidth(10)
  
print(f"Lambda value used for Transformation: {fitted_lambda}")
```
